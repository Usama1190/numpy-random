```python
from numpy import random

x = random.normal(size=(2, 6))

print(x)
```

```python
from numpy import random

x = random.normal(loc=3, scale=3, size=(2, 3))

print(x)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.displot(random.normal(size=(1000)), kind='kde')

plt.show()
```

```python
from numpy import random

x = random.binomial(n=20, p=0.5, size=100)

print(x)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.displot(random.binomial(n=20, p=0.5, size=10))

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    "normal": random.normal(loc=50, scale=5, size=1000),
    "binomial": random.binomial(n=20, p=0.5, size=1000)
}

sns.displot(data, kind='kde')

plt.show()
```
